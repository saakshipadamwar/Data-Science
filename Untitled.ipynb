```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv('Iris.csv')
```

```python
iris.head()
```

```python
features = iris.columns[1:-1]
```

```python
iris.info()
```

```python
for feature in features:
    plt.figure(figsize=(10,5))
    plt.hist(x = iris[feature], bins = None)
```

```python
for feature in features:
    plt.figure(figsize=(10,5))
    sns.boxplot(x = iris[feature], y = iris['Species'])
```
